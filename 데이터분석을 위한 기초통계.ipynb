## 매년 달라지는 수확량을 그래프로 분석하자.
- 수확량은 300kg에서 700kg 사이에서 분포하고 있다.

## 수확량의 평균을 구해보자.
$$
{ 466kg + 605kg + 387kg + 406kg + 555kg + 554kg + 572kg + 634kg + 623kg + 534kg \over 10} = 533.6kg
$$

## 매년 수확량이 평균에서 얼마나 떨어져 있는지 확인해보자.
$$
표준편차: s_x = \sqrt{\frac{1}{n}\sum(x_i - \bar{x})^2} = 82.37
$$

## 수확량에 영향을 주는 변수 정의
- 마을의 인구수, 농부가 소유한 논의 넓이, 올해 내린 비의 양 등

image.png

### 독립변수 & 종속변수
- 독립변수란, 종속변수에 영향을 주는 변수이다.
- 종속변수란, 독립변수를 조절할 때 그에 따라서 변화되는 변수이다.

image.png

예를 들어, 농부가 비료의 양(독립변수)을 조절하면 수확량(종속변수)이 변하는 것입니다.

image.png

## 상관관계
- 비료의 양과 수확량 간의 상관관계를 분석해보니, 비료를 많이 줄수록 수확량이 증가하는 경향이 있었습니다.

image.png

### 상관관계가 곧 인과관계는 아니다!
- 예를 들어, 달-토성간 거리와 물리학 학사 졸업생수 사이에도 강력한 상관관계가 있습니다.
- 이때 만약 지구상에 더 많은 물리학 학사 졸업생들이 증가한다고 하더라도 달-토성간 거리가 변하지는 않는다.

즉, 상관관계는 있지만 인과관계는 아니라는 뜻이다.

image.png

## 인과관계를 확인하기 위해서 가설을 세우자
- `비료의 양을 20% 늘리면 수확량이 15% 증가할 것이다`라는 가설을 세우자.

image.png

- 가설검정을 위해 실험을 진행했지만, 결과가 항상 일관되지는 않았습니다.
- 즉 어떤 때는 가설과 일치했고, 어떤 때는 그렇지 않았습니다.

image.png

### p-값
- 이런 불확실성을 줄이기 위해서 p-값이 필요합니다.
- p-값은 우리의 가설이 틀력다고 가정할 때, 관찰된 결과(또는 그보다 더 극단적인 결과)가 우연히 발생할 확률을 나타냅니다.

image.png

- p-값이 너무 작을 경우, 통상 0.05보다 작을 경우, 귀무가설을 기각합니다.

image.png

- 만약 95% 이상 확률로 비료가 수확량 증가에 영향을 준다면, 귀무가설을 기각하게 된다.
- 즉 비료의 양과 수확량은 서로 인과관계라고 할 수 있게 된다.

image.png
